#Why Semantic search ?
---

Used to Identifying the most relevant and useful data from the huge corpus is very important for business to improve customer satisfaction. AI and machine learning has contributed to empower the content searching.

#What is Semantic search ?
---
Semantic Search means searching the words in the data-store based on their meanings. It is a searching attempt made to generate accurate results by understanding the searcher intent, query context and the relationship between words but not just focusing on one string or literal matching.

```python
from langchain_community.document_loaders import PyPDFLoader

#loader = PyPDFLoader("Semantic_Search.pdf")
loader = PyPDFLoader("https://link.springer.com/content/pdf/10.1007/0-387-34456-X_52.pdf")
pages = loader.load_and_split()
```

#Chunking for semantic search: -

Chunking is the process of breaking down the large document into smaller segments.

Idea:
RecursiveCharacterTextSplitter(chunk_size, chunk_overlap) Chunk_Overlap = 0 or 1.

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts= text_splitter.split_documents(pages)
print(f"After chunking: {len(texts)} document(s)")
```

```python
print(texts)
```

#Embeddings :-

Embedding is technique of converting text into numerical form. Vector is the resultant of embeddings. Based on the type of data like audio, video, image, a 3-layered Neural network is used to create vector embeddings

```python
#Wrapper around sentence_transformers embedding models.
#To use, you should have the sentence_transformers python package installed.
from langchain.embeddings import HuggingFaceEmbeddings
from tqdm.autonotebook import tqdm
#Instantiate Hugging face embeddings
embeddings = HuggingFaceEmbeddings()
```

```python
print(embeddings)
```

#Vector database :-

Storing and indexing vector database :

#How vector database works?

Vector databases works based on 3 steps namely:

1. Generating vector embeddings

2. Indexing

3. Query processing


Input(Images, text/words,Audio, video) ->3 Layer Neural network ->embdeddings ->Vector Database[]


#Storing and indexing vector database

```python
from langchain.vectorstores import FAISS
database=FAISS.from_documents(texts,embeddings)
```

#Query matching

```python
#User query
query = "What is semantics search ?"
#similarity search on the query using the FAISS Vector DB
docs= database.similarity_search(query)
docs
```

#LLMs Models for synthesizing the response

```python
from getpass import getpass
import os
from langchain import HuggingFaceHub
HUGGINGFACEHUB_API_TOKEN = getpass()
os.environ["HUGGINGFACEHUB_API_TOKEN"] = HUGGINGFACEHUB_API_TOKEN
FM_model = "tiiuae/falcon-7b"
llm = HuggingFaceHub(repo_id=FM_model, model_kwargs={"temperature":0.6, "max_length":64})
```

```python
#Langchain Questions and Answering library
from langchain.chains.question_answering import load_qa_chain
#Load Q&A chain with LLM- Falcon
chain=load_qa_chain(llm,chain_type='stuff')
query="Wnat is semantic search"
#Apply similarty search
docs= database.similarity_search(query)
#Run the chain
chain.run(input_documents=docs, question =query)
```

#Haystack features:

Haystack is an open-source framework used for building search systems.

building blocks:

1. Nodes (Text Processing)
2. Pipelines (Reader and retriever)

 a)Reader : Analyse the documents for Q & A tasks

 b)Retriever : number of documents to retrieve
3. Agents (Empowered by LLM to decide which action to take)
4. REST API (Add memory to remember history of interactions)
